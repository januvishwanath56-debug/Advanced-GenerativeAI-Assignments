```python
# Unique Words in a Sentence

sentence = input("Enter sentence: ")

words = sentence.split()
unique_words = set(words)

print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

```python
# Highest Salary from Employee Data

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

max_emp = ""
max_sal = 0

for name, sal in employees.items():
    if sal > max_sal:
        max_sal = sal
        max_emp = name

print("Highest Salary:", max_emp, "-", max_sal)
```

```python
# Find Maximum and Minimum Values

numbers = [45, 22, 89, 10, 66]

mx = numbers[0]
mn = numbers[0]

for n in numbers:
    if n > mx:
        mx = n
    if n < mn:
        mn = n

print("List:", numbers)
print("Max:", mx)
print("Min:", mn)
```

```python
# Count Products Above a Price Threshold

prices = [450, 1200, 899, 1500, 300]

count = 0

for p in prices:
    if p > 1000:
        count += 1

print("Products above 1000:", count)

```

```python
# Calculate Attendance Percentage

attendance = ["P", "P", "A", "P", "P"]

present = 0

for a in attendance:
    if a == "P":
        present += 1

percentage = (present / len(attendance)) * 100

print("Attendance Percentage:", percentage)
```

```python
# Remove Duplicate Phone Numbers

phone_numbers = [9876543210, 9123456789, 9876543210]

unique_nums = set(phone_numbers)

print("Unique phone numbers:", unique_nums)
```

```python
# Count Character Frequency

text = "python"

freq = {}

for ch in text:
    if ch in freq:
        freq[ch] += 1
    else:
        freq[ch] = 1

print(freq)
```

```python
# Convert List to Tuple

lst = [10, 20, 30]

t = tuple(lst)

print("Tuple:", t)
```

```python
# Check if a Key Exists in a Dictionary

emp = {"Ravi": 75000, "Anita": 68000}

key = "Ravi"

if key in emp:
    print("Employee exists")
else:
    print("Employee not found")
```

```python
# Calculate Average Marks

marks = [70, 80, 65, 75]

total = 0

for m in marks:
    total += m

avg = total / len(marks)

print("Average Marks:", avg)
```

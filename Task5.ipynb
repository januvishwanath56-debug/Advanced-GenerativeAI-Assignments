```python
# Social Media Post Engagement Analyzer

def analyze_engagement(likes):
    total = 0
    for like in likes:
        total += like

    status = "Viral Post" if total >= 1000 else "Normal Engagement"

    print("Total Likes:", total)
    print("Post Status:", status)

likes_list = [200, 300, 250, 300]
analyze_engagement(likes_list)
```

```python
# Medicine Stock Alert System

def medicine_stock(stock):
    status = "Low Stock Alert" if stock < 10 else "Stock Sufficient"

    print("Medicine Stock:", stock)
    print("Status:", status)

medicine_stock(6)
```

```python
# Rainfall Adequacy Checker

def rainfall_checker(rainfall, required):
    total = 0
    for r in rainfall:
        total += r

    avg = total / len(rainfall)
    status = "Adequate Rainfall" if avg >= required else "Inadequate Rainfall"

    print("Average Rainfall:", avg)
    print("Rainfall Status:", status)

rainfall_checker([70, 75, 68, 75], 70)
```

```python
# Duplicate Account Detection

def check_duplicates(usernames):
    unique_users = set(usernames)

    if len(unique_users) != len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")

users = ["john", "alex", "john", "mike"]
check_duplicates(users)
```

```python
# Appointment Eligibility Checker

def appointment_eligibility(age):
    status = "Eligible" if age >= 18 else "Not Eligible"

    print("Patient Age:", age)
    print("Eligibility Status:", status)

appointment_eligibility(21)
```

```python
# Premium Crop Price Filter

def premium_crops(prices):
    premium = []

    for price in prices:
        if price > 2000:
            premium.append(price)

    print("Premium Crops:", premium)

premium_crops([1500, 2500, 1800, 3200])
```

```python
# Application Health Checker

def system_health(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print("Error Count:", errors)
    print("System Status:", status)

system_health(7)
```

```python
# Daily Transaction Limit Checker

def transaction_checker(amount):
    status = "Approved" if amount <= 50000 else "Rejected"

    print("Transaction Amount:", amount)
    print("Transaction Status:", status)

transaction_checker(60000)
```

```python
# Student Attendance Eligibility System

def attendance_eligibility(attendance):
    present = 0

    for day in attendance:
        if day == "P":
            present += 1

    percentage = (present / len(attendance)) * 100
    status = "Eligible" if percentage >= 75 else "Not Eligible"

    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)

attendance_eligibility(["P","P","A","P","P"])
```

```python
# Smart Electricity Bill Analyzer

def electricity_bill(units):
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    print("Total Bill:", bill)
    print("Usage Level:", usage)

electricity_bill(250)
```

```python
# User Login Check

username = "admin"
password = "1234"

#get data from user

u = input("Enter username: ") 
p = input("Enter password: ")

if u == username and p == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
# Pass / Fail Analyzer

marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for m in marks:
    if m >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("Pass:", pass_count)
print("Fail:", fail_count)
```

```python
# Simple Data Cleaner

names = [" Alice ", "bob", " CHARLIE "]

cleaned = []

for n in names:
    cleaned.append(n.strip().lower())

print(cleaned)
```

```python
# Message Length Analyzer

messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
    l = len(msg)
    print(f"{msg}={l}")
    if l > 10:
        print(f"{msg} = Long message")
```

```python
# Error Message Detector

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

count = 0

for log in logs:
    if log == "ERROR":
        count += 1

print("Errors:", count)
```

```python
# Q1: Smart Parking Lot Management System

def smart_parking(capacity, logs):
    parked = 0
    peak = 0

    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1

        if parked > peak:
            peak = parked

    print("Currently Parked Vehicles:", parked)

    if parked > capacity:
        print("Parking Status: Full - Capacity Exceeded")
    else:
        print("Parking Status: Available")

    print("Peak Parking Usage:", peak)


# Sample Input
smart_parking(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

```python
# Q2: Online Food Delivery Time Estimator

def estimate_delivery_time(distance, traffic, weather):

    time = distance * 5   # base delivery time

    if traffic == "High":
        time += 10
    elif traffic == "Medium":
        time += 5

    if weather == "Rainy":
        time += 5
    elif weather == "Stormy":
        time += 15

    print("Estimated Delivery Time:", time, "minutes")


# Sample Input
estimate_delivery_time(8, "High", "Rainy")
```

```python
# Q3: Movie Theatre Seat Occupancy Analyzer

def seat_occupancy(total_seats, booked_seats):

    booked_count = len(booked_seats)
    occupancy = (booked_count / total_seats) * 100

    print("Occupancy:", round(occupancy), "%")

    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
        print("Suggestion: Open additional shows")
    else:
        print("Show Status: Seats Available")


# Sample Input
seat_occupancy(200, [1]*150)
```

```python
# Q4: Cloud Server Load Classification System

def classify_server_load(cpu_readings):

    avg_load = sum(cpu_readings) / len(cpu_readings)

    print("Average CPU Load:", round(avg_load), "%")

    if avg_load < 50:
        status = "Normal"
    elif 50 <= avg_load <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print("Server Status:", status)


# Sample Input
classify_server_load([45, 60, 70, 85, 90])
```

```python
# Q5: Smart Classroom Resource Usage Monitor

def resource_monitor(resources):

    overused = []

    for resource, hours in resources.items():
        if hours > 8:   # overuse threshold
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input
resource_monitor({
    "Projector": 6,
    "AC": 9,
    "Lights": 4
})
```

```python
# Q6: Online Event Registration Capacity Controller

def event_registration(capacity, registrations):

    if registrations <= capacity:
        confirmed = registrations
        waitlist = 0
        status = "Open"
    else:
        confirmed = capacity
        waitlist = registrations - capacity
        status = "Closed"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status:", status)


# Sample Input
event_registration(100, 105)
```

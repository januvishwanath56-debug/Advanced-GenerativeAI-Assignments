```python
# Mobile Recharge Validation System

def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: ₹"))

        if amount < 50:
            print("Recharge must be at least ₹50. Try again.")
            continue

        if amount in valid_plans:
            print("Recharge Successful!")
            break
        else:
            print("Invalid plan. Valid plans are:", valid_plans)

validate_recharge()
```

```python
# Inventory Reorder Alert System

def inventory_alert(products):
    for item, stock in products.items():
        if stock < 15:
            print(item, ": Reorder Alert")
        else:
            print(item, ": Stock OK")

products = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 8
}

inventory_alert(products)
```

```python
# Student Result Processing System

def student_result(marks):
    total = 0
    for m in marks:
        total += m

    avg = total / len(marks)
    return "Pass" if avg >= 50 else "Fail"

marks = [60, 70, 40, 55]
print(student_result(marks))
```

```python
# Cab Fare Estimator

def calculate_fare(km, peak):
    fare = 50 + 12 * km
    if peak == "yes":
        fare += fare * 0.25
    return fare

distance = float(input("Enter KM: "))
peak = input("Peak hour? yes/no: ")
print("Fare =", calculate_fare(distance, peak))
```

```python
# Employee Attendance Eligibility Checker

def attendance_checker(att):
    present = 0
    for d in att:
        if d == "P":
            present += 1

    percent = (present / len(att)) * 100
    return "Eligible" if percent >= 75 else "Not Eligible"

records = ["P","A","P","P","P"]
print(attendance_checker(records))
```

```python
# Password Strength Checker

def check_password(pwd):
    digit = False
    special = False

    for ch in pwd:
        if ch.isdigit():
            digit = True
        if ch in "@#$":
            special = True

    if len(pwd) >= 8 and digit and special:
        return "Strong"
    return "Weak"

password = input("Enter password: ")
print(check_password(password))
```
